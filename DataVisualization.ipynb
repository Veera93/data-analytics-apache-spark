```python
import os 
```

```python
def getdata(file_name, issorted):
    file_name = os.path.dirname(os.getcwd())+"/data-analytics-apache-spark/Output/"+file_name+".txt"
    f  = open(file_name, "r")
    data = f.read() 
    lines = data.split("\n")
    dict = {}
    for line in lines:
        if(line != ''):
            key, value = line.split("\t")
            dict[key] = int(value)
    sorted_dic = sorted(dict.items(), key=lambda x:x[1],reverse=True)
    if issorted == True:
        return sorted_dic
    else:
        return dict
```

```python
def getWordCount(inputData):
    dic= []
    for item in inputData :
        pair = {}
        pair["name"] = item[0]
        pair["weight"]= item[1]
        dic.append(pair)
    return dic
    
```

```python
business = getdata("Business", True)
business = getWordCount(business)
```
